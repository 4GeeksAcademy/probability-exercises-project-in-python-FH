## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np

def tirada_de_dados(n_simulación = 1000):
    conteo = 0
    
    # Cada iteración del bucle for es una prueba.
    for i in range(n_simulación):
        
        # Tirada de cada Dado
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        
        # Suma los valores para obtener la puntuación
        puntuación = dado1 + dado2
        
        # Decide si debemos agregarlo al conteo
        if (puntuación % 2 == 0) or (puntuación > 7):
            conteo += 1

    # Calculate probability
    return conteo / n_simulación
print(f"La probabilidad de obtener un número par o mayor que 7 es: {np.round(tirada_de_dados() * 100, 2)}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
Caja_Bolas = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        Caja_Bolas[i] = "Blanca"
    elif (i > 9) and (i < 30):
        Caja_Bolas[i] = "Roja"
    else:
        Caja_Bolas[i] = "Verde"

print(Caja_Bolas)
            
# TODO
def tomar_bolas(n_simulación = 1000):
    conteo_1 = 0
    conteo_2 = 0

    for i in range(n_simulación):
        colores = []

        # Take 5 balls from the box
        for i in range(5):
            colores.append(Caja_Bolas[np.random.randint(0, 59)])

        # Convert list to Numpy array for better filtering
        colores = np.array(colores)
        
        Bolas_Blancas = sum(colores == "Blanca")
        Bolas_Rojas = sum(colores == "Roja")
        Bolas_Verdes = sum(colores == "Verde")

        # Decide if we should add it to the count:
        if (Bolas_Blancas == 3) and (Bolas_Rojas == 2):
            conteo_1 += 1
        
        if (Bolas_Blancas == 5) or (Bolas_Rojas == 5) or (Bolas_Verdes == 5):
            conteo_2 += 1
    
    return conteo_1 / n_simulación, conteo_2 / n_simulación

probabilidad = tomar_bolas(1000)

print(f"La probabilidad de que salgan 3 blancas y 2 rojas es: {np.round(probabilidad[0] * 100, 2)}%")
print(f"La probabilidad de que todas sean del mismo color es: {np.round(probabilidad[1] * 100, 2)}%")

```
